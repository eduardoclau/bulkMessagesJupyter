```python
import sqlite3
import time
import urllib
import webbrowser

import pyautogui as pg


def criar_tabela():
    conn = sqlite3.connect('contatos.db')
    c = conn.cursor()
    c.execute('''CREATE TABLE IF NOT EXISTS contatos
                 (numero TEXT)''')
    conn.commit()
    conn.close()


conn = sqlite3.connect('contatos.db')
conn.close()

criar_tabela()


def adicionar_numero():
    novo_numero = input("Digite o número a ser adicionado:")
    conn = sqlite3.connect('contatos.db')
    c = conn.cursor()
    c.execute(f"INSERT INTO contatos VALUES ('{novo_numero}')")
    conn.commit()
    conn.close()
    return "Contato adicionado."


def remover_numero():
    remover_numero = input("Digite o número a ser removido:")
    conn = sqlite3.connect('contatos.db')
    c = conn.cursor()
    c.execute(f"DELETE FROM contatos WHERE numero = '{remover_numero}'")
    conn.commit()
    conn.close()
    return "Contato removido."


def enviar_mensagem():
    msg = input("Digite a mensagem a ser enviada:")
    msg_formatted = msg.replace("\\n", "\n") # substitui os caracteres '\n' por quebra de linha
    # codifica a mensagem para ser usada na URL
    msg_encoded = urllib.parse.quote_plus(msg_formatted)

    conn = sqlite3.connect('contatos.db')
    c = conn.cursor()
    c.execute("SELECT numero FROM contatos")
    numeros = c.fetchall()
    for numero in numeros:
        webbrowser.open_new_tab(
            f'https://web.whatsapp.com/send?phone=+55\
                {numero[0]}&text={msg_encoded}')
        time.sleep(5)  # aguarda 10 segundos para carregar a página do WhatsApp
        pg.press('enter')
    conn.close()
    return "Mensagem enviada."



while True:
    opcao = input(
        "Digite 1 para adicionar contato:   \
            Digite 2 para remover contato:  \
            Digite 3 para enviar mensagem:  \
            Digite 4 para sair: ")

    if opcao == "1":
        adicionar_numero()

    elif opcao == "2":
        remover_numero()

    elif opcao == "3":
        enviar_mensagem()

    elif opcao == "4":
        break

    else:
        print("Opção inválida, tente novamente.")
```
